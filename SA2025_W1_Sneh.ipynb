### 1. What is name of car that has the highest horsepower?

### 2. How many cars have mpg ≥ 35?

```python
#your code here
no_cars = grand_prix[grand_prix['mpg']>=35]['mpg'].count()
no_cars
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
most_common_origin = grand_prix[(grand_prix['horsepower'] > 100) & (grand_prix['weight'] < 3000)]['origin'].mode()[0]
print(most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
mean_acc = grand_prix[grand_prix['origin'] == 'japan']['acceleration'].mean().round(2)
mean_acc
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median = grand_prix.mpg.median()
grand_prix = grand_prix.reset_index()
best_hp_to_weight = grand_prix.name.loc[grand_prix.hp_per_weight[grand_prix.mpg > median].idxmax()]
print('The car with best hp to weight and above-median mpg is', best_hp_to_weight)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg = grand_prix.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('model_Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data = grand_prix,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    palette='deep',
    sizes=(40, 400),
    alpha=0.7
)

plt.title('Horsepower vs Weight by Origin (Size = MPG)')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
